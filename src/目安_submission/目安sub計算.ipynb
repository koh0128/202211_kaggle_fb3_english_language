```python
COMPETION_DATA_DIR = '/content/drive/MyDrive/ColabNotebooks/kaggle/english_language/目安_submission/'
```

```python
sub1 = pd.read_csv(COMPETION_DATA_DIR + 'submission_1.csv')
sub2 = pd.read_csv(COMPETION_DATA_DIR + 'submission_2.csv')
sub3 = pd.read_csv(COMPETION_DATA_DIR + 'submission_3.csv')
sub4 = pd.read_csv(COMPETION_DATA_DIR + 'submission_4.csv')
sub5 = pd.read_csv(COMPETION_DATA_DIR + 'submission_5.csv')

sub1
```

```python
pred_sub = pd.DataFrame()
pred_sub['text_id'] = sub1['text_id']
for col_nm in ['cohesion', 'syntax', 'vocabulary', 'phraseology', 'grammar', 'conventions']:
    pred_sub[col_nm] = (sub1[col_nm] + sub2[col_nm] + sub3[col_nm] + sub4[col_nm] + sub5[col_nm]) / 5 
pred_sub 
```

```python
pred_sub.to_csv(COMPETION_DATA_DIR + 'submission_pred.csv')
```
